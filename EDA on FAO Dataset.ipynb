```python
import numpy as np
import pandas as pd
data_df = pd.read_csv("fooddata2.csv") 
data_df.to_csv("fooddata2.csv", index = False)
```

```python
#Show description of the dataset
data_df.head()
```

```python
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively? Ans: 209460.54 and 269617.53
data_df.groupby("Item")["Y2014", "Y2017"].agg([sum])
```

```python
#What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places? Ans = 135.236 and 1603.404 respectively
meandata = data_df["Y2015"].mean()
standarddev = data_df["Y2015"].std()
print(round(meandata,3))
print(round(standarddev,3))
```

```python
#What is the total number and percentage of missing data in 2016 to 2 decimal places? 1536 and 2.52 respectively
print(round(data_df.isna().sum(),2),data_df.isna().mean()*100)
```
